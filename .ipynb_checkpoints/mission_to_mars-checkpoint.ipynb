# Web Scraping Mars Information


```python
# Dependencies
from bs4 import BeautifulSoup as Bs
import requests
# import os # Man, was this giving wonky errors which we were only able to attribute to Linux. :(

from splinter import Browser
executable_path = {"executable_path": "/bin/chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)

url = "https://mars.nasa.gov/news/"
browser.visit(url)

# Scrape the browser into soup and use soup to find the full resolution image of mars
html = browser.html
soup = Bs(html, 'html.parser')
```

```python
title = soup.select("#page > div.grid_list_page.module.content_page > div > article > div > section > div > ul > li:nth-child(1) > div > div > div.content_title > a")[0].text
```

## JPL Mars Space Images - Featured Image

```python
url2 = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
browser.visit(url2)

# Scrape the browser into soup and use soup to find the full resolution image of mars
html2 = browser.html
soup2 = Bs(html2, 'html.parser')
```

```python
soup2.find_all(class_="button")[0]
# .click()
```

```python
xpath2 = '//*[@id="full_image"]'
```

```python
results = browser.find_by_xpath(xpath2)
img = results[0]
img.click()
```
